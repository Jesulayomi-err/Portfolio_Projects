```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/global_meat_production.csv')
df.head(5)
```

```python
df.rename(columns={'Meat, total | 00001765 || Production | 005510 || tonnes':'Total_production(tonnes)'}, inplace=True)
df.head(5)
```

```python
df.describe()
```

```python
df.groupby('Country')['Total_production(tonnes)'].sum().sort_values(ascending=False).head(20)
```

```python
#Top 5 most populous countries
top5_populated_countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan']
```

```python
#China
country= df['Country']==top5_populated_countries[0]
china= df[country].groupby(['Year','Country'])['Total_production(tonnes)'].sum().sort_values(ascending=False)
print(china)
```

```python
#India
country=df['Country']==top5_populated_countries[1]
india=df[country].groupby(['Year','Country'])['Total_production(tonnes)'].sum().sort_values(ascending=False)
print(india)
```

```python
#United states
country=df['Country']==top5_populated_countries[2]
usa=df[country].groupby(['Year','Country'])['Total_production(tonnes)'].sum().sort_values(ascending=False)
print(usa)
```

```python
#Indonesia
country=df['Country']==top5_populated_countries[3]
indonesia=df[country].groupby(['Year','Country'])['Total_production(tonnes)'].sum().sort_values(ascending=False)
print(indonesia)
```

```python
#Pakistan
country=df['Country']==top5_populated_countries[4]
pakistan=df[country].groupby(['Year','Country'])['Total_production(tonnes)'].sum().sort_values(ascending=False)
print(pakistan)
```

```python
#Global meat and dairy production over time
 (df[['Year', 'Total_production(tonnes)']][df.Country == 'World'].groupby('Year').sum() / 1000000).plot(kind='line', ylabel='Million Tons', y='Total_production(tonnes)') #Fixed typo in the column name from Total_producition(tonnes) to Total_production(tonnes)
```

```python
years = np.array(df['Year'].values)
china_production = np.array(china[years])
india_production = np.array(india[years])
us_production = np.array(usa[years])
indonesia_production = np.array(indonesia[years])
pakistan_production = np.array(pakistan[years])
```

```python
plt.figure(figsize = (15, 12))
plt.plot(years, china_production, label = 'china')
plt.plot(years, india_production,label= 'india')
plt.plot(years, us_production, label = 'usa')
plt.plot(years, indonesia_production, label = 'indonesia')
plt.plot(years, pakistan_production, label = 'pakistan')
plt.title('Meat & Dairy production in Top 5 most populous countries')
plt.xlabel('Years')
plt.ylabel('Production in Tonnes')
plt.grid(True)
plt.legend()
plt.show()
```

There are varoius factors that can hinder constant supply of these produce, many of which are interconnected and have both local and global implications. Some of the key reasons include:

**Seasonal Variations**
Livestock Production Cycles: Meat and dairy production are influenced by the breeding and growth cycles of livestock. During the dry season, livestock may have less access to grazing lands and water, reducing production and leading to price increases.
Feed Availability: During certain times of the year, the availability of feed for livestock decreases, raising the cost of animal care, which in turn affects meat and dairy prices.

**Supply Chain Disruptions**
Logistics and Transportation Issues: Poor infrastructure, especially in rural areas, increases transportation costs. Delays or damage during transport can cause losses, contributing to price hikes.
Insecurity: In regions affected by conflict or banditry, livestock farming may be disrupted, leading to reduced supply and increased prices of meat and dairy.

**Climatic Conditions**
Drought and Flooding: Nigeria has experienced extreme weather patterns, including drought and flooding. These affect pasturelands and water availability for livestock, which directly impacts meat and dairy production.
Desertification in the North: The encroaching desert reduces the availability of grazing lands, particularly in northern Nigeria, where many livestock are reared.

**Animal Disease Outbreaks**
Diseases like Foot-and-Mouth Disease: Outbreaks of diseases among livestock can lead to a significant drop in meat and dairy production. These diseases not only reduce the availability of healthy livestock but also increase costs related to vaccination and treatment.

**Government Policies and Regulation**
Import Restrictions: Nigeria has put restrictions on the import of meat and dairy to encourage local production. While this can boost local industries in the long term, it often leads to short-term price increases due to limited supply and production capacity.
Subsidies and Incentives: Fluctuations in government subsidies for livestock farmers, whether for feed, veterinary services, or fuel, can cause cost fluctuations.

**Economic Factors**
Inflation and Currency Fluctuations: The devaluation of the Naira against major currencies affects the cost of imported feed, veterinary supplies, and equipment, driving up the cost of meat and dairy products.
Fuel Prices: The cost of fuel impacts transportation, production, and the overall cost of livestock farming. Any increases in fuel prices will likely cause an increase in the prices of meat and dairy.

**Demand and Consumption Patterns**
Festive Seasons: Demand for meat, particularly during Muslim festivals like Eid al-Fitr and Eid al-Adha, or Christmas, can lead to temporary spikes in prices due to increased consumption.
Urbanization and Changing Diets: As urbanization increases, demand for meat and dairy products in urban areas grows. This demand is often not matched by increased production, contributing to price fluctuations.


**Herdsmen-Farmer Conflicts**
Displacement of Farmers and Livestock: Conflicts between herdsmen and farmers, particularly in the Middle Belt region, disrupt livestock farming, decrease production, and limit the supply of both meat and dairy.


Understanding the various factors that affects the production of meat and dairy products in Nigeria, stake holders can address the issues being faced in the supply chain in Nigeria and find ways to reduce the volatility of the market.
